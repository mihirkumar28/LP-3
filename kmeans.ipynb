#**Unsupervised Learning**
Unsupervised machine learning is where the underlying structure of the data is suitably identified. It is used when the dataset is not labelled that is the output/target value is not known. Machine Learning algorithms are used to find hidden patterns or data groupings. Its ability to discover similarities and differences in information make it the ideal solution for exploratory data analysis, cross-selling strategies, customer segmentation, and image recognition. 

Unsupervised learning models are used for three main tasks: clustering, association and dimensionality reduction:
- Clustering is a technique for grouping unlabeled data based on their similarities or differences. For example, K-means clustering algorithms assign similar data points into groups, where the K value represents the size of the grouping and granularity. This technique is helpful for market segmentation, image compression, etc.
- Association is another type of unsupervised learning method that uses different rules to find relationships between variables in a given dataset. These methods are frequently used for market basket analysis and recommendation engines, along the lines of Customers Who Bought This Item Also Bought recommendations.
- Dimensionality reduction is a learning technique used when the number of features  (or dimensions) in a given dataset is too high. It reduces the number of data inputs to a manageable size while also preserving the data integrity. Often, this technique is used in the preprocessing data stage, such as when autoencoders remove noise from visual data to improve picture quality.

Mathematically, unsupervised learning for any given set $X$ aims to learn the underlying distribution of the data which can be leveraged for various applications that is the algorithm aims to estimate a model/heuristic which can represent the probability distribution for new input $x_{t}$ given previous inputs $x_{1},...,x_{t-1}$ which is nothing but $P(x_{t}|x_{1},...,x_{t-1})$.

#**Clustering**
The main intuition behind clustering is that for a given set of points (instances) we wish to group similar instances and when a new instance is encountered we wish to assign it to the correct group. It groups unlabeled data based on their similarities or differences. Clustering algorithms are used to process raw, unclassified data objects into groups represented by structures or patterns in the information.

$\large f_{clustering}: X_{i} \rightarrow C_{i}$,  

$X_{i}$ $\large \epsilon$ $\mathbb{R^{m}}$, 

$C_{i}$: Cluster  $\large \epsilon$  $C$, where $C$ is the set of all clusters. 

The number of groups/clusters into which the data can be grouped is unknown and variable. 


#**Clustering vs Classification**
- Classification is a supervised learning task whereas clustering is an unsupervised learning task.
- Classification models a heuristic function which enables an arbitary input $X_{i}$ to be assigned a class $y$ from the set of classes available $Y$. 
    
    $\large f_{classification}: X_{i} \rightarrow y$, $X_{i}$ $\large \epsilon$ $\mathbb{R^{m}}$ 
- Clustering on the other hand aims to group instances into clusters $C_{i}$ such that similar instances will belong to the same the cluster $C_{i}$  $\large \epsilon$  $C$, where $C$ is the set of all clusters. 

    $\large f_{clustering}: X_{i} \rightarrow C_{i}$, $X_{i}$ $\large \epsilon$ $\mathbb{R^{m}}$ 
- Clustering works with unlabelled data whereas classification works with labelled data.
- In classification task, number of classes are known. In clustering, number of groups are unknown.

#**K-means**
##**Bloom's Taxonomy (Understand & Remember)**.
K-means algorithm is the most popular and yet simplest of all the clustering algorithms. It is an exclusive clustering method that is a given instance $X_{i}$ will only belong to one cluster $C_{i}$.

For grouping purposes, a given number of centroids $m_{i}$ are calculated and the distance of the given instance $X_{i}$ from such centroids is calculated. $X_{i}$ is assigned to that group whose centroid is closest to it. 

After an instance has been assigned to a cluster, centroids are recalculated. 

k-means.png

This algorithm is non-deterministic in nature as two seperate iterations with same input may produce different outputs.

k-means iteration.webp

##**Quality Metric**
The quality of the cluster assignments is determined by computing the sum of the squared error (SSE) after the centroids converge, or match the previous iterations assignment. The SSE is defined as the sum of the squared Euclidean distances of each point to its closest centroid. Since this is a measure of error, the objective of k-means is to try to minimize this value.

##**Reasoning for non-deterministic nature**
The initialization of the centroids is an important step. The random initialization step causes the k-means algorithm to be nondeterministic, meaning that cluster assignments will vary if one runs the same algorithm twice on the same dataset.

##**Number of iterations**
After choosing a number of clusters and the initial centroids, the expectation-maximization step is repeated until the centroid positions reach convergence and/or are unchanged.

##**k-value**
###**Bloom's Taxonomy (Evaluate)**
Note that solving this problem of finding optimal value exactly is NP-hard. Intuitively then, the optimal choice of k will strike a balance between maximum compression of the data using a single cluster, and maximum accuracy by assigning each data point to its own cluster

Value of k in k-means can be determined by:
- **Elbow Methods**: 
One should choose a number of clusters so that adding another cluster doesn't give much better modeling of the data. More precisely, if one plots the Sum of Squared Errors against the number of clusters, the first clusters will have high error values, but at some point the marginal gain will drop, giving an angle in the graph.



Calculate the **Within-Cluster-Sum of Squared Errors (WSS)** for different values of k, and choose the k for which WSS becomes first starts to diminish. In the plot of WSS-versus-k, this is visible as an elbow.

##**WSS**
- The Squared Error for each point is the square of the distance of the point from its representation i.e. its predicted cluster center.
- The WSS score is the sum of these Squared Errors for all the points.
- Any distance metric like the Euclidean Distance or the Manhattan Distance can be used.
- The idea here is that a given instance should be as close to its centroid as possible. For given set of clusters $C$ ={$C_{1},...,C_{k}$}, $c_{i}$ is the centroid for cluster $C_{i}$

$$ f_{WSS} = \sum_{i=1}^k\sum_{C_i~\in~C} ||x_i - c_i||^{2} \rightarrow \min\limits_C,$$
It is to be noted that this function would be minimum when number of clusters equals the number of instances that is each instance is a cluster in itself, which is degerate case as it leads to no findings and hence should be avoided. As seen in the elbow method described earlier, that number is chosen as the number of clusters after which a sharp decline in WSS is not observed.

For k-means clustering, follow the steps in the given order:
- Arbitrarily choose an initial $k$ centers $C$ = {$c_{1}$, $c_{2}$,    , $c_{k}$}. It is standard practice to choose the initial centers uniformly at random from $X$
- For each $i$ $\large \epsilon$ {$1$, . . . , $k$}, set the cluster $C_{i}$ to be the set of points in $X$ that are closer to $c_{i}$ than they are to $c_{j}$ for all $j$ $\neq$ $i$.
- For each $i$ $\large \epsilon$ {$1$, . . . , $k$}, set $c_{i}$ to be the centeroid of all points in $C_{i}$: $c_{i}$ =$|Ci|^{-1}$ $\large \sum_{xC_{i}}$ $x$.
-  Repeat Steps 2 and 3 until $C$ no longer changes
- For $X  R^{d}$, choose $k$ centers $m_{i}$ so as to minimize the potential function

    $$ =$\large \sum_{x  X} $ $\min_{cC}$ $||x-c||^{2}$
.

- The idea here is that Steps 2 and 3 are both guaranteed to decrease $$, so the algorithm makes local improvements to an arbitrary clustering until it is no longer possible to do so.

```python
def cal_dis(point1,point2):
  return norm(np.array(point1)-np.array(point2))
```

```python
def random_select(data,k):
    C=random.sample(list(data1.values()),k)
    return C
```

##**Bloom's Taxonomy (Apply)**

```python
def kmeans(data,k):
  keys=list(data.keys())
  points=list(data.values())
  C=random_select(points,k)
  wcss=0
  cluster={str(i):[] for i in range(k)}
  while True:
    for point in points:
      dist=[]
      for centroid in C:
        dist.append(cal_dis(point,centroid))
      min_dist=min(dist)
      cluster[str(dist.index(min_dist))].append(point)
      wcss=+(min_dist)**2
    new_C=[]
    for i in cluster:
      new_C.append([float(sum(j))/len(j) for j in zip(*cluster[i])])
    if  sorted(C) == sorted(new_C):
      return cluster, C, wcss, k
    else: C = new_C
```

```python
def optimum_k(data):
  partial_func=functools.partial(kmeans,data)
  queue=[]
  with concurrent.futures.ThreadPoolExecutor() as executor:
    for i in range(1,len(data.keys())):
      queue.append(executor.submit(partial_func,i))
    for thread in concurrent.futures.as_completed(queue):
      print(thread.result()[2],thread.result()[3])
```

```python
optimum_k(data1)
```

```python
from sklearn.cluster import KMeans
data2=pd.DataFrame(list(data1.values()))
wcss = []
for i in range(1, 9):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(data2)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 9), wcss,marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```
