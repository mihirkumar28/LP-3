#**Supervised Learning**
Supervised machine learning is where heuristic is selected/learnt which allows mapping of input to outputs based on previously seen instances.
On observing the data, it can be seen that the dataset is labelled that is input and output values are distinctly known.  
Mathematically, 
For a finite sequence $S$=$((x_{1},y_{1}),...,(x_{n},y_{n}))$ of pairs from $X$ x $Y$. $y_{i}$ $\large \epsilon$ $Y$ is called the label corresponding $x_{i} \large \epsilon X$. This is the input.

The output is a hypothesis $h$ such that $\large h: X \rightarrow Y$ that aims at predicting $y$ $\large \epsilon$ $Y$ for arbitary $x$ $\large \epsilon$ $X$.

#**k-NN**
It is an instance based learning method. The basic idea behind it is that every instance will look like its neighbours that is instances that are closer together in the instance space will have similar features and thus, will belong to the same class. It is based on the hypothesis that if the distances are measured well enough, similar instances will belong to the same class.

To classify any instance $X_{new}$ :

- Find distance of  $X_{new}$ from all $X_{i}$ $\large \epsilon$ $X_{data}$, i $\large \epsilon$ $\mathbb{N}$ 
- Pick k nearest neighbours that is k instances closest to $X_{new}$
- The class of $X_{new}$ will be same as that of the majority of selected neighbours. Mathematically, we find the probabilty of instances belonging to a particular class from the chosen k nearest neighbours.

     $P(y=j|X=x)$=$|k|^{-1} \large \sum_{i \epsilon A} I(y^{i}=j)$

A: Set of k points that are closest to x   
I(x) is the indicator function which evaluates to 1 when the argument x is true and 0 otherwise.

#**Learning in k-NN**
K-Nearest Neighbours is an Instance based learning method. Instance based learning methods are also called 'lazy learners'. Unlike other learning methods, here a target function/model is not developed for the entire instance space. At runtime, for every unseen instance, a target function is approximated locally and differently for each such unseen instance. Instance-based learning means that our algorithm doesn’t explicitly learn a model. Instead, it chooses to memorize the training instances which are subsequently used as “knowledge” for the prediction phase. Concretely, this means that only when a query to our database is made (i.e. when we ask it to predict a label given an input), will the algorithm use the training instances to predict an answer.

kNN is  non-parametric means it makes no explicit assumptions about the functional form of h, avoiding the dangers of mismodeling the underlying distribution of the data. For example, suppose our data is highly non-Gaussian but the learning model we choose assumes a Gaussian form. In that case, our algorithm would make extremely poor predictions.

#**k value**:
Choosing appropriate k value is very critical in the k-NN algorithm 
- A small value of k means that noise will have a higher influence on the result. 
- Larger values of K will have smoother decision boundaries which mean lower variance but increased bias, it is also computationally expensive and defeats the basic philosophy behind KNN ( that points that are near might have similar densities or classes ).
- A simple approach to select k is set **k = sqrt(N)**, where N is the size of the dataset
- Another way to choose k is though cross-validation, by selecting the k value which performs best on the validation set and minimizes the validation error.
- Generally **odd values of k** are preferred in order to avoid a split vote.

#**Performance Metrics**
The quality of classification/regression with k-NN depends on several parameters:

- The number of neighbors k
- The distance measure between samples (common ones include Hamming, Euclidean, cosine, and Minkowski distances). Note that most of these metrics require data to be scaled. 
-Weights of neighbors (each neighbor may contribute different weights; for example, the further the sample, the lower the weight).

#**Why k-NN (Analyze)**
- Simple to implement and intuitive to understand
- Can learn non-linear decision boundaries when used for classfication and regression. Can came up with a highly flexible decision boundary adjusting the value of K
- **No Training Time for classification/regression** : The KNN algorithm has no explicit training step and all the work happens during prediction
- **Constantly evolves with new data**: Since there is no explicit training step, as we keep adding new data to the dataset, the prediction is adjusted without having to retrain a new model. Makes it robust to noisy data as well.
- **Choice of distance metric**: There are many distance metrics to chose from. Some popular distance metrics used are Euclidean, Manhattan, Minkowski, hamming distance eand so on

#**k-NN vs Linear Algorithms & Decision Trees**
k-NN works well for tasks where data is linearly inseperable that is data points of different classes cannot be suitably seperated by a line unlike linear algorithms.


Decision Trees are used for classification purposes. However, the heuristic approximated by decision trees is some cases is too complex. The negative impact of noisy data on decision trees is significant. k-NN approximates a complex target function by means of locally estimated less complex heuristics and is also robust to noisy data.

```python
import concurrent.futures
import functools
import copy
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt
```

```python
data={
    'x':[2, 4, 4, 4, 6, 6],
    'y':[4, 2, 4, 6, 2, 4],
    'label':["orange", "orange", "blue", "orange", "blue", "orange"]
}
```

```python
plt.scatter(data['x'],data['y'],c=data['label'])
plt.grid(True)
plt.show()
```

```python
def cal_dis(point,key_point):
  return norm(np.array(point)-np.array(key_point))
```

#**Bloom's Taxonomy (Apply)**

```python
def kNN(key_point,data,k):
  data_list=list(data.values())
  ls=[]
  for (i,j,counter) in zip(data_list[0],data_list[1],range(len(data_list[0]))):
    dist=cal_dis([i,j],key_point)
    ls.append([dist,counter])
  ls.sort()
  #print(ls)
  val=0
  for i in range(k):
    if data_list[2][ls[i][1]] == 'blue': val=val+1
    else: val=val-1
    #print(val)
  if val>0: return 'blue',k
  return 'orange',k
```

```python
def optimum_k(data):
  data1=copy.deepcopy(data)
  key_point=[data1['x'][0],data1['y'][0]]
  label=data1['label'][0]
  for item in data1.values():
    item.remove(item[0])
  partial_func=functools.partial(kNN,key_point,data1)
  queue=[]
  with concurrent.futures.ThreadPoolExecutor() as executor:
    for i in range(1,len(data1['x'])+1,2):
      queue.append(executor.submit(partial_func,i))
  for thread in concurrent.futures.as_completed(queue):
    print('For k value:',thread.result()[1],'\nLabel predicted:',thread.result()[0],'\nOriginal label:',label)
    if thread.result()[0] == label:
      print('Correctly labelled\n')
    else:
      print('Incorrectly labelled\n')
```

#**Bloom's Taxonomy (Evaluate)**

```python
optimum_k(data)
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics

df=pd.DataFrame(data)
labels=df['label'] 
df=df.drop(columns=['label'])
X_train, X_test, y_train, y_test = train_test_split(df,labels, test_size=0.2,random_state=3)
for i in range(1,len(X_train)+1,2):
  knn = KNeighborsClassifier(n_neighbors=i)
  model=knn.fit(X_train, y_train)
  y_pred = knn.predict(X_test)
  print('For k=',i,'accuracy=',metrics.accuracy_score(y_test, y_pred))
```
